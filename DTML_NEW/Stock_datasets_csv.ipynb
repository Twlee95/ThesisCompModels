## stock csv reader

```python
import pandas as pd
import numpy as np
import torch.nn as nn
import torch
import os

class stock_csv_read:
    def __init__(self, data, x_frames, y_frames):
        self.data = data
        self.x_frames = x_frames
        self.y_frames = y_frames
        self.stock_data = self.data_loader()
        self.index_data = self.index_data_loader()
        
    def data_loader(self):
        stock_data = pd.read_csv("./data/kdd17/ourpped/"+self.data,header=None)
        stock_data = stock_data[29:]
        stock_data = stock_data.iloc[:,:12]
        target = pd.DataFrame(np.where(stock_data.iloc[:,11]>=0,1,0))
        stock_data = stock_data.iloc[:,:11].reset_index(drop=True)
        stock_data = pd.concat([stock_data,target],axis=1 , ignore_index= True)
        
        return stock_data
    
    def index_data_loader(self):
        stock_data = pd.read_csv("./data/kdd17/ourpped1/IXIC.csv", header=None)
        return stock_data

    def spliter(self, data_list):
        self.dd = data_list
        length_ = len(self.dd[0])
        
        split_list = []
        for i in range(length_-self.x_frames-self.y_frames+1):
            index_stock_list = []
            for stock_index in self.dd:    # 첫번째꺼는 주식 두번째꺼는 idnex
                xy = []
                X = stock_index.iloc[i : i +self.x_frames, 0:11].values
                y = stock_index.iloc[i +self.x_frames : i +self.x_frames +self.y_frames, 11:].values
                xy.append(X)
                xy.append(y)
                index_stock_list.append(xy)
            split_list.append(index_stock_list)
        return split_list
    
    def cv_split(self):
        stock_data = self.stock_data
        index_data = self.index_data
        
        data_len = len(stock_data)
        mok = data_len//19
        
        adder = 0
        data_list = []
        for i in range(10):
            sp_data = stock_data.iloc[0+adder:10*mok+adder,:]
            
            index_sp_data = index_data.iloc[0+adder:10*mok+adder,:]
            
            tvt = [] ## train validation test가 분할된 리스트 목록 리스트
            
            ## 개별주가 분할
            train_sp_data = sp_data[0:8*mok]
            validation_sp_data = sp_data[8*mok:9*mok]
            test_sp_data = sp_data[9*mok:10*mok]
            
            ## index 분할
            train_sp_index = index_sp_data[0:8*mok]
            validation_sp_index = index_sp_data[8*mok:9*mok]
            test_sp_index = index_sp_data[9*mok:10*mok]
            
            # 개별주가, index list
            train_list = [train_sp_data, train_sp_index]
            validation_list = [validation_sp_data, validation_sp_index]
            test_list = [test_sp_data, test_sp_index]

            # blocked cross validation 분할 및 10:1 분할
            train_sp_data_ = self.spliter(train_list)
            validation_sp_data_ = self.spliter(validation_list)
            test_sp_data_ = self.spliter(test_list)

            tvt.append(train_sp_data_)
            tvt.append(validation_sp_data_)
            tvt.append(test_sp_data_)
            
            adder += mok
            data_list.append(tvt)
            
        return data_list
```
